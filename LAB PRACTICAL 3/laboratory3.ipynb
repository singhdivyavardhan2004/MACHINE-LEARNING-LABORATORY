```python
import pandas as pd

data = {
    "Outlook": ["Sunny","Sunny","Overcast","Rain","Rain","Rain","Overcast","Sunny","Sunny","Rain","Sunny","Overcast","Overcast","Rain"],
    "Temp": ["Hot","Hot","Hot","Mild","Cool","Cool","Cool","Mild","Cool","Mild","Mild","Mild","Hot","Mild"],
    "Humidity": ["High","High","High","High","Normal","Normal","Normal","High","Normal","Normal","Normal","High","Normal","High"],
    "Wind": ["Weak","Strong","Weak","Weak","Weak","Strong","Strong","Weak","Weak","Weak","Strong","Strong","Weak","Strong"],
    "Play": ["No","No","Yes","Yes","Yes","No","Yes","No","Yes","Yes","Yes","Yes","Yes","No"]
}

df = pd.DataFrame(data)
df
```

```python
X = {
    "Outlook":"Sunny",
    "Temp":"Cool",
    "Humidity":"High",
    "Wind":"Strong"
}
```

```python
yes = df[df["Play"] == "Yes"]
no = df[df["Play"] == "No"]

print("Yes samples =", len(yes))
print("No samples =", len(no))
```

```python
def likelihood(feature,value,cls):
    return len(cls[cls[feature]==value]) / len(cls)
```

```python
P_yes = len(yes) / len(df)
P_no = len(no) / len(df)

print("P(Yes) =", P_yes)
print("P(No)  =", P_no)
```

```python
def probability(feature, value, data):
    return len(data[data[feature] == value]) / len(data)
```

```python
X = {
    "Outlook": "Sunny",
    "Temp": "Cool",
    "Humidity": "High",
    "Wind": "Strong"
}
```

```python
P_X_yes = P_yes
P_X_no = P_no

for feature, value in X.items():
    P_X_yes *= probability(feature, value, yes)
    P_X_no *= probability(feature, value, no)

print("P(Yes | X) =", P_X_yes)
print("P(No | X)  =", P_X_no)
```

```python
if P_X_yes > P_X_no:
    print("Prediction: Play = Yes")
else:
    print("Prediction: Play = No")
```
